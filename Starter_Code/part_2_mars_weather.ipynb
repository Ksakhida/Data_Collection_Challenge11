```python
# Import relevant libraries
from splinter import Browser
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import pandas as pd
```

```python
browser = Browser('chrome')
```

### Step 1: Visit the Website

Use automated browsing to visit the [Mars Temperature Data Site](https://static.bc-edx.com/data/web/mars_facts/temperature.html). Inspect the page to identify which elements to scrape.

   > **Hint** To identify which elements to scrape, you might want to inspect the page by using Chrome DevTools to discover whether the table contains usable classes.


```python
# Visit the website
# https://static.bc-edx.com/data/web/mars_facts/temperature.html
url = "https://static.bc-edx.com/data/web/mars_facts/temperature.html"
browser.visit(url)
```

### Step 2: Scrape the Table

Create a Beautiful Soup object and use it to scrape the data in the HTML table.

Note that this can also be achieved by using the Pandas `read_html` function. However, use Beautiful Soup here to continue sharpening your web scraping skills.

```python
# Create a Beautiful Soup Object
html = browser.html
soup = BeautifulSoup(html, 'html.parser')

#Display soup
soup
```

```python
# Extract all rows of data
all_rows = soup.find_all('tr', class_="data-row")

#Print all rows of data
all_rows
```

### Step 3: Store the Data

Assemble the scraped data into a Pandas DataFrame. The columns should have the same headings as the table on the website. Here’s an explanation of the column headings:

* `id`: the identification number of a single transmission from the Curiosity rover
* `terrestrial_date`: the date on Earth
* `sol`: the number of elapsed sols (Martian days) since Curiosity landed on Mars
* `ls`: the solar longitude
* `month`: the Martian month
* `min_temp`: the minimum temperature, in Celsius, of a single Martian day (sol)
* `pressure`: The atmospheric pressure at Curiosity's location

```python
# Create an empty list
list_rows = []

# Loop through the scraped data to create a list of rows
for row in all_rows:
    td = row.find_all('td')
    row = [column.text for column in td]
    list_rows.append(row)

#Print the list of rows
list_rows
```

```python
# Create a Pandas DataFrame by using the list of rows and a list of the column names
mars_df = pd.DataFrame(list_rows, columns = ["id", "terrestrial_date", "sol", "ls", "month", "min_temp", "pressure"])
```

```python
# Confirm DataFrame was created successfully
mars_df
```

### Step 4: Prepare Data for Analysis

Examine the data types that are currently associated with each column. If necessary, cast (or convert) the data to the appropriate `datetime`, `int`, or `float` data types.

  > **Hint** You can use the Pandas `astype` and `to_datetime` methods to accomplish this task.


```python
# Examine data type of each column
mars_df.dtypes
```

```python
# Change data types for data analysis
mars_df['terrestrial_date'] = pd.to_datetime(mars_df['terrestrial_date'])
mars_df['sol'] = mars_df['sol'].astype('int')
mars_df['ls'] = mars_df['ls'].astype('int')
mars_df['month'] = mars_df['month'].astype('int')
mars_df['min_temp'] = mars_df['min_temp'].astype('float')
mars_df['pressure'] = mars_df['pressure'].astype('float')
```

```python
# Confirm type changes were successful by examining data types again
mars_df.dtypes
```

### Step 5: Analyze the Data

Analyze your dataset by using Pandas functions to answer the following questions:

1. How many months exist on Mars?
2. How many Martian (and not Earth) days worth of data exist in the scraped dataset?
3. What are the coldest and the warmest months on Mars (at the location of Curiosity)? To answer this question:
    * Find the average the minimum daily temperature for all of the months.
    * Plot the results as a bar chart.
4. Which months have the lowest and the highest atmospheric pressure on Mars? To answer this question:
    * Find the average the daily atmospheric pressure of all the months.
    * Plot the results as a bar chart.
5. About how many terrestrial (Earth) days exist in a Martian year? To answer this question:
    * Consider how many days elapse on Earth in the time that Mars circles the Sun once.
    * Visually estimate the result by plotting the daily minimum temperature.


```python
# 1. How many months are there on Mars?
mars_months = mars_df["month"].value_counts().sort_index()

#Print number of months
mars_months
```

```python
#Additional question in the instructions - How many months exist on Mars
# Number of months on Mars
total_months = mars_df['month'].nunique()

#Print total number of months on Mars
total_months
```

```python
# 2. How many Martian days' worth of data are there?
martian_days = mars_df['sol'].nunique()

#Print Martian days
martian_days
```

```python
# 3. What is the average low temperature by month?
avg_min_temp_by_month = mars_df.groupby('month')['min_temp'].mean()

#Print average low temperature by month
avg_min_temp_by_month
```

```python
# Plot the average temperature by month
avg_min_temp_by_month.plot(kind='bar', title='Average Minimum Temperature by Month')
plt.ylabel('Temperature in Celsius')
plt.xlabel('Month')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
#What are the coldest and the warmest months on Mars (at the location of Curiosity)? To answer this question:
#Find the average the minimum daily temperature for all of the months.
#Plot the results as a bar chart.


#Firstly lets determine the coldest and warmest months on Mars(at the location of Curiosity)
coldest_month = avg_min_temp_by_month.idxmin()
warmest_month = avg_min_temp_by_month.idxmax()
print(f"On average, the coldest month on Mars is month {coldest_month}, and the warmest month is month {warmest_month}.")

# Identify the coldest and hottest months in Curiosity's location
coldest_hottest_months = avg_min_temp_by_month.sort_values()

#Print coldest and hottest months in Curiosity's location
coldest_hottest_months

# Plotting the average temperatures
coldest_hottest_months.plot(kind='bar', title='Coldest and Hottest Temperature by Months')
plt.ylabel('Temperature in Celsius')
plt.xlabel('Month')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# 4. Average pressure by Martian month
avg_pressure_by_month = mars_df.groupby('month')['pressure'].mean()

#Print average pressure by Martian month
avg_pressure_by_month
```

```python
#Which months have the lowest and the highest atmospheric pressure on Mars? To answer this question:
#Find the average the daily atmospheric pressure of all the months.
#Plot the results as a bar chart.

#lets determine the months for the lowest and highest atmospheric pressure on Mars
lowest_pressure_month = avg_pressure_by_month.idxmin()
highest_pressure_month = avg_pressure_by_month.idxmax()
print(f"On average, the month with the lowest atmospheric pressure on Mars is month {lowest_pressure_month}, and the highest pressure is in month {highest_pressure_month}.")

#Identify (by sorting) the lowest and the higest atmospheric pressure on Mars
lowest_highest_pressure = avg_pressure_by_month.sort_values()

#Print the lowest and the higest atmospheric pressure on Mars 
lowest_highest_pressure

#Plot the average pressure by month
lowest_highest_pressure.plot(kind='bar', title='Average Pressure by Month')
plt.ylabel('Atmospheric Pressure')
plt.xlabel('Month')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# About how many terrestrial (Earth) days exist in a Martian year? To answer this question:
# Consider how many days elapse on Earth in the time that Mars circles the Sun once.
# Visually estimate the result by plotting the daily minimum temperature.

# 5. How many terrestrial (earth) days are there in a Martian year?

#Converting terrestrial_date to datetime
mars_df['terrestrial_date'] = pd.to_datetime(mars_df['terrestrial_date'])

# Sort the mars dataframe by terrestrial date
mars_df = mars_df.sort_values(by='terrestrial_date')

#Plot daily minimum temperature
plt.figure(figsize=(8, 5))
mars_df['min_temp'].plot()
plt.xlabel('Number of terrestrial days')
plt.ylabel('Minimum temperature (°C)')
plt.title('Daily Minimum Temperature Over Terrestrial Days')
plt.grid
plt.show()

# After visualizing the plot, identify peaks and estimate the number of days between them
# Manually identify the approximate day indices for two consecutive peaks
# This numbers are identified from the plot visually
peak1_day = 750  
peak2_day = 1425  

# Calculate the estimated Martian year length in terrestrial days
mars_year_estimate = peak2_day - peak1_day

print(f"A visual estimate of a Martian year length is approximately {mars_year_estimate} terrestrial days.")

# Known value from external sources (Internet) for comparison
mars_year_days = 687  

print(f"Internet search confirms that a Mars year is equivalent to {mars_year_days} Earth days.")


```

On average, the third month has the coldest minimum temperature on Mars, and the eighth month is the warmest. But it is always very cold there in human terms!



Atmospheric pressure is, on average, lowest in the sixth month and highest in the ninth.

The distance from peak to peak is roughly 1425-750, or 675 days. A year on Mars appears to be about 675 days from the plot. Internet search confirms that a Mars year is equivalent to 687 earth days.
